```python
data = {'Nome': ['Nicolas','Daniel','Rodolfo'],
        'Idade': [25,29,24],
        'Cidade': ['São Paulo','Ibaté','São Carlos']}
```

```python
#Dataframe é o formato comom o Pandas trabalha com os dados. Dataframe PySpark
```

```python
#Transforma um dicionário ou lista em um data frame
df = pd.DataFrame(data)
```

```python
df[['Nome','Idade']].isna().sum()
```
